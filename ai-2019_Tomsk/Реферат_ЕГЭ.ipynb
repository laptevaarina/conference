## Анализ первой части ЕГЭ

```python
#функция перевода "+" и "-" в 1 и 0
def signsInDigits(s):
    new_s = ''
    for i in s:
        if i == '+':
            new_s += '1'
        elif i == '-':
            new_s += '0'
        else:
            new_s += i
    A = [int(i) for i in new_s]
    return A
```

```python
#создаём новые DataFrame для наглядности
new_rus = pd.DataFrame(columns = [str(i+1) for i in range(24)])
new_math = pd.DataFrame(columns = [str(i+1) for i in range(12)])
```

```python
#переводим значения для русского языка
for i, j in enumerate(rus['Задания с кратким ответом']):
    digits = signsInDigits(j)
    new_rus.loc[i] = [k for k in digits]

new_rus.insert(0,'Фамилия',rus['Фамилия'][:])

#добавляем сумму первой части ЕГЭ по русскому языку
summs = []
for row in new_rus.itertuples():
    summa = 0
    for j in range(2, len(row)):
        summa += int(row[j])
    summs.append(summa)

new_rus.insert(25, 'summa', summs)

#добавляем сумму по заданиям первой части ЕГЭ по русскому языку 
sum_rus = []
for i in range(1,25):
    sum_rus.append(new_rus[str(i)].sum())
new_rus.loc[33] = ['Сумма по заданиям'] + [j for j in sum_rus] + [' ']
new_rus
```

```python
#переводим значения для математики
for i, j in enumerate(math['Задания с кратким ответом']):
    digits = signsInDigits(j)
    new_math.loc[i] = [k for k in digits]

new_math.insert(0,'Фамилия',math['Фамилия'][:])

#добавляем сумму первой части ЕГЭ по математике
summs1 = []
for row1 in new_math.itertuples():
    summa1 = 0
    for j in range(2, len(row1)):
        summa1 += int(row1[j])
    summs1.append(summa1)

new_math.insert(13, 'summa', summs1)

#добавляем сумму по заданиям первой части ЕГЭ по математике
sum_math = []
for i in range(1,13):
    sum_math.append(new_math[str(i)].sum())
new_math.loc[63] = ['Сумма по заданиям'] + [j for j in sum_math] + [' ']
new_math
```

## Анализ второй части ЕГЭ

```python
def points(s):
    summa2 = ''
    for i in range(0, len(s), 4):
        summa2 += s[i]
    S = [int(k) for k in summa2]
    return S
```

##### Русский язык

```python
#разбиваем строчку на разные ячейки
rus_2 = pd.DataFrame(columns = [str(i+1) for i in range(12)])
for i, j in enumerate(rus['Задания с развёрнутым ответом']):
    digits2 = points(j)
    rus_2.loc[i] = [k for k in digits2]

rus_2.insert(0,'Фамилия',rus['Фамилия'][:])

#подсчитываем сумму баллов второй части у каждого выпускника
summs2 = []
for row in rus_2.itertuples():
    summa = 0
    for j in range(2, len(row)):
        summa += int(row[j])
    summs2.append(summa)

rus_2.insert(13, 'summa', summs2)

#подсчитываем сумму по заданиям второй части
sum_rus2 = []
for i in range(1,13):
    sum_rus2.append(rus_2[str(i)].sum())
rus_2.loc[33] = ['Сумма по заданиям'] + [j for j in sum_rus2] + [' ']

rus_2.head()
```

##### Математика

```python
#разбиваем строчку на разные ячейки
math_2 = pd.DataFrame(columns = [str(i+1) for i in range(7)])
for i, j in enumerate(math['Задания с развёрнутым ответом']):
    digits2 = points(j)
    math_2.loc[i] = [k for k in digits2]

math_2.insert(0,'Фамилия',math['Фамилия'][:])

#подсчитываем сумму баллов второй части у каждого выпускника
summs2 = []
for row in math_2.itertuples():
    summa = 0
    for j in range(2, len(row)):
        summa += int(row[j])
    summs2.append(summa)

math_2.insert(8, 'summa', summs2)

#подсчитываем сумму по заданиям второй части
sum_math2 = []
for i in range(1,8):
    sum_math2.append(math_2[str(i)].sum())
math_2.loc[64] = ['Сумма по заданиям'] + [j for j in sum_math2] + [' ']

math_2.head()
```
